```python
import math

def calc_f1_score(cp, tp, fn):
    if (not isinstance(tp, int)) or (not isinstance(cp, int)) or (not isinstance(fn, int)):
        if not isinstance(tp, int):
            print("tp must be int")
        if not isinstance(cp, int):
            print("cp must be int")
        if not isinstance(fn, int):
            print("fn must be int")
        return

    if tp <= 0 or cp <= 0 or fn <= 0:
        print("tp and cp and fn must be greater than zero")
        return

    precision = tp / (tp + cp)
    recall = tp / (tp + fn)
    f1_score = (2 * precision * recall) / (precision + recall)

    print('precision:', precision)
    print('recall:', recall)
    print('f1_score:', f1_score)

    return f1_score
```

```python
calc_f1_score(5, 4, 8)
```

```python
calc_f1_score(1.2, 4, 8)
```

```python
import math

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

def activation_function(x, activation_function_name):
    alpha = 0.01

    if not is_number(x):
        print('x must be a number')
        return

    x = float(x)

    if activation_function_name not in ['sigmoid', 'relu', 'elu']:
        print(f'{activation_function_name} is not supported')
        return

    if activation_function_name == 'sigmoid':
        result = 1 / (1 + math.exp(-x))
        print(f'{activation_function_name}: f({x}) = {result}')
    elif activation_function_name == 'relu':
        result = max(0, x)
        print(f'{activation_function_name}: f({x}) = {result}')
    elif activation_function_name == 'elu':
        result = x if x > 0 else alpha * (math.exp(x) - 1)
        print(f'{activation_function_name}: f({x}) = {result}')

# Test function
x = input("Nhập giá trị x: ")
activation_function_name = input("Nhập tên activation function (sigmoid, relu, elu): ")

activation_function(x, activation_function_name)
```

```python
import numpy as np

def calculate_loss(num_samples, loss_name):
    if not num_samples.isnumeric():
        print("Number of samples must be an integer number")
        return

    num_samples = int(num_samples)
    if num_samples < 1:
        print("Number of samples must be a positive integer")
        return

    # Tạo các mẫu ngẫu nhiên cho predict và target
    sample = np.arange(num_samples)
    predict = np.random.uniform(0, 10, num_samples)
    target = np.random.uniform(0, 10, num_samples)

    # Tính các giá trị loss
    if loss_name == "MAE":
        loss = np.mean(np.abs(predict - target))
    elif loss_name == "MSE":
        loss = np.mean((predict - target) ** 2)
    elif loss_name == "RMSE":
        loss = np.sqrt(np.mean((predict - target) ** 2))
    else:
        print("Invalid loss name. Choose either 'MAE', 'MSE', or 'RMSE'.")
        return

    # In các giá trị
    print("Loss name:", loss_name)
    print("Sample:", sample)
    print("Predict:", predict)
    print("Target:", target)
    print("Loss:", loss)

# Ví dụ sử dụng:
num_samples = input("Nhập số lượng samples: ")
loss_name = input("Nhập tên loss (MAE, MSE, RMSE): ")

calculate_loss(num_samples, loss_name)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def approx_sin(x, n):
    sin_x = 0
    for i in range(n):
        term = ((-1)**i * x**(2*i + 1)) / factorial(2*i + 1)
        sin_x += term
    return sin_x

def approx_cos(x, n):
    cos_x = 0
    for i in range(n):
        term = ((-1)**i * x**(2*i)) / factorial(2*i)
        cos_x += term
    return cos_x

def approx_sinh(x, n):
    sinh_x = 0
    for i in range(n):
        term = (x**(2*i + 1)) / factorial(2*i + 1)
        sinh_x += term
    return sinh_x

def approx_cosh(x, n):
    cosh_x = 0
    for i in range(n):
        term = (x**(2*i)) / factorial(2*i)
        cosh_x += term
    return cosh_x
```

```python
import numpy as np

def md_nre(y, y_hat, n, p):
    """
    Tính Mean Difference of n-th Root Error (MD_nRE)

    Args:
    - y (list hoặc numpy array): Giá trị thực tế.
    - y_hat (list hoặc numpy array): Giá trị dự đoán.
    - n (int): Bậc của căn.
    - p (int): Bậc của giá trị lỗi.

    Returns:
    - MD_nRE (float): Giá trị MD_nRE.
    """
    y = np.array(y)
    y_hat = np.array(y_hat)

    m = len(y)
    error_sum = 0.0

    for i in range(m):
        error_sum += np.abs(y[i]**(1/n) - y_hat[i]**(1/n))**p

    md_nre_value = (1/m) * error_sum

    return md_nre_value

# Ví dụ sử dụng:
y = [0.5, 0.9, 0.1, 0.4, 0.3]
y_hat = [0.6, 0.8, 0.2, 0.5, 0.4]
n = 2
p = 1

print(f"MD_nRE: {md_nre(y, y_hat, n, p)}")
```
